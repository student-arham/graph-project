```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#LINE PLOT
x=np.linspace(0,10,100)
y=np.sin(x)
```

```python
plt.plot(x,y,color='green')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('A Line Plot of sin(X)')

plt.grid()
plt.show()
```

```python
# SCATTER PLOT
x=np.random.rand(20)
y=np.random.rand(20)
```

```python
plt.scatter(x,y,color='blue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('A Scatter Plot')

plt.grid()
plt.show()
```

```python
# BAR CHART
score=[75,90,100,85,50,95]
name=['Alan','Ken','Sam','Alice','Jane','Ann']
```

```python
plt.bar(name,score,color='grey')
plt.xlabel('Student Name')
plt.ylabel('Student Score')
plt.title('Exam Score')

plt.grid()
plt.show()
```

```python
#HISTOGRAM PLOT
data=np.random.rand(1000)
plt.hist(data,bins=40,color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Value Frequency')

plt.grid()
plt.show()
```

```python
#PIE CHART
score=[75,90,100,85,50,95]
name=['Alan','Ken','Sam','Alice','Jane','Ann']
```

```python
plt.pie(score,labels=name,colors=["red","yellow","orange","brown","blue","pink"])
plt.title("Pie Chart")

plt.show()
```

```python
# Heatmap (Correlation matrix)
df = pd.DataFrame(np.random.rand(5,5), columns=['A','B','C','D','E'])
sns.heatmap(df,annot=True,cmap='coolwarm')
plt.title("Seaborn Heatmap")

plt.show()
```
